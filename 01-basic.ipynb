```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np

start = 0; end = 10; amount = 25
x = np.linspace(start, end, amount) # arreglo de numeros
y = x * x
# print(np.array([x,y]).reshape(25, 2))
```

```python
pylab.plot(x, y, 'r')
```

```python
pylab.subplot(1,2,1) # (rows, columns, index)

pylab.plot(x,y, 'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0.5, 0.1, 0.8, 0.8]) # left, bottom, width, height of canvas (from 0 to 1)

axes.plot(x,y, 'r.')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2) # submap is of 1 row, 2 cols

for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# Graph inside another
fig = plt.figure()
                    # Lef, Bot, Wid, Hei of canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small axes

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize=(16,9), dpi=300)

fig.add_subplot()
plt.plot(x,y, 'r')
```

```python
fig, axes = plt.subplots()
axes.set_title("title")
axes.set_xlabel("X Label")
axes.set_ylabel("y Label")
axes.plot(x, x**2, 'r')
axes.plot(x, x**3, 'b')

axes.legend(["y = x**2", "y = x**3"], loc=2)
```

```python
fig, axes = plt.subplots(dpi = 150)
axes.plot(x, x**1.2, color='red', alpha=.8)
axes.plot(x, x+2, color='#1155dd', alpha=.8, linestyle='--')
axes.plot(x, x+3, alpha=.5, lw=3)

line, = axes.plot(x, x+4, c="black", lw=1.5)
line.set_dashes([5, 10, 15, 10])

axes.plot(x, x+5, marker='o', markersize=3)
```

```python
fig, Axes = plt.subplots(1, 2, figsize=(10,5)) # 1 row, 2 cols

Axes[0].plot(x, x**2, 'r.', x, x**3, 'g.', lw=2)
Axes[0].grid(True)

Axes[1].plot(x, x**2, 'r.-', x, x**3, 'g.-', lw=2)
Axes[1].set_ylim(0, 60)
Axes[1].set_xlim(2, 5)
```
