```python
# bar plot
res = sns.barplot(mtcars['cyl'], mtcars['mpg'])
plt.show()
```

```python
res = sns.barplot(x='cyl', y='mpg', data= mtcars, color='blue')
plt.show()
```

```python
# Countplot

sns.countplot(x = 'cyl', data = mtcars, palette="Set1")
```

```python
sns.countplot(x = 'carb', data = mtcars, palette="rocket")
```

```python
sns.countplot(x='gear', hue='carb', data=mtcars, palette='rocket')
```

```python
# Displot

sns.distplot(mtcars.mpg, bins = 10, color = 'g')
```

```python
# Heatmap | Correlations between variables

sns.heatmap(mtcars.corr(), cbar = True, linewidths = 0.5)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data=iris)
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data=iris, hue = 'species', alpha=0.7)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette="Set1")
```

```python
# lmplot

sns.lmplot(x = 'petal_length', y = 'petal_width', data = iris, hue='species', markers=['o','*','+'])
```

```python
# Boxplot

sns.boxplot(x = 'species', y = 'petal_length', data = iris)
```

```python
sns.jointplot(x = "petal_width", y="petal_length", data=iris, hue='species', height=5)
```
