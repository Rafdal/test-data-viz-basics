```python
from matplotlib import pylab
import numpy as np
from matplotlib import pyplot as plt

start = 0; end = 10; amount = 25
x = np.linspace(start, end, amount) # arreglo de numeros
y = x * x
```

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(16,5))

axes[0].set_title('scatter plot')
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)) )

axes[1].set_title('step plot')
axes[1].step(n, n**2, lw=2)

axes[2].set_title('bar plot')
axes[2].bar(n, 1+n**2, lw=1, alpha=0.5)

axes[3].set_title('fill_between')
axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)

t = np.linspace(0, 2*np.pi, 100) ## de 0 a 2pi  con 100 puntos
ax.plot(t,t, c='blue', lw=3)
```

```python
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].set_title("Default histogram")
axes[0].hist(n, bins=15)

axes[1].set_title("Cumulative detailed histogram")
# axes[1].hist(n, cumulative=False, bins=100, alpha=0.5)
axes[1].hist(n, cumulative=True, color='red', bins=100, alpha=0.5)
```

```python
# Contour image

import matplotlib.cm as cm

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp( -X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2 )

Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, levels=10)
ax.clabel(CS, inline=1, fontsize = 10)
ax.set_title('contour map')
```
